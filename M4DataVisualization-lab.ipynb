```python
# print all the tables names in the database
QUERY = """
SELECT name as Table_Name FROM
sqlite_master WHERE
type = 'table'
"""
# the read_sql_query runs the sql query and returns the data as a dataframe
pd.read_sql_query(QUERY,conn)
```

```python
QUERY = """
SELECT Age,COUNT(*) as count
FROM master
group by age
order by age
"""
pd.read_sql_query(QUERY,conn)
```

# Hands-on Lab


## Visualizing distribution of data


### Histograms


Plot a histogram of `ConvertedComp.`


```python
# your code goes here
# Fetch all data from the master table
QUERY = """
SELECT *
FROM master
"""

# Load the data into a DataFrame
df = pd.read_sql_query(QUERY, conn)
import matplotlib.pyplot as plt

# Ensure that the ConvertedComp column is in numeric format and drop NaN values
df['ConvertedComp'] = pd.to_numeric(df['ConvertedComp'], errors='coerce')

# Plot the histogram of the ConvertedComp column
plt.figure(figsize=(10, 6))
plt.hist(df['ConvertedComp'].dropna(), bins=30, edgecolor='black')
plt.title('Histogram of Converted Compensation (Annual Salaries)')
plt.xlabel('Converted Compensation (USD)')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the box plot of the Age column
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Age'])
plt.title('Box Plot of Age')
plt.xlabel('Age')
plt.grid(axis='x')
plt.show()

```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
# Plot the scatter plot of Age and WorkWeekHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['WorkWeekHrs'], alpha=0.5)
plt.title('Scatter Plot of Age vs. Work Week Hours')
plt.xlabel('Age')
plt.ylabel('Work Week Hours')
plt.grid()
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
# Plot the bubble plot of WorkWeekHrs and CodeRevHrs
plt.figure(figsize=(10, 6))
plt.scatter(df['WorkWeekHrs'], df['CodeRevHrs'], 
            s=df['Age']*10,  # Bubble size based on Age (scaled for visibility)
            alpha=0.5, 
            edgecolors='w',  # Optional: add white edges to bubbles for visibility
            linewidth=0.5)

plt.title('Bubble Plot of Work Week Hours vs. Code Review Hours')
plt.xlabel('Work Week Hours')
plt.ylabel('Code Review Hours')
plt.grid()
plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Filter for age group 30 to 35
filtered_df = df[(df['Age'] >= 30) & (df['Age'] <= 35)]

# Calculate median WorkWeekHrs and CodeRevHrs
median_hours = filtered_df[['WorkWeekHrs', 'CodeRevHrs']].median()

# Create a DataFrame for plotting
median_hours_df = median_hours.reset_index()
median_hours_df.columns = ['Hours Type', 'Median Hours']

# Plotting
plt.figure(figsize=(8, 5))
plt.bar(median_hours_df['Hours Type'], median_hours_df['Median Hours'], color=['blue', 'orange'])
plt.title('Median WorkWeekHrs and CodeRevHrs (Age 30 to 35)')
plt.xlabel('Hours Type')
plt.ylabel('Median Hours')
plt.xticks(rotation=0)

# Display the value on top of the bars
for index, value in enumerate(median_hours_df['Median Hours']):
    plt.text(index, value, round(value, 2), ha='center', va='bottom')

plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Filter for age group 45 to 60
filtered_df = df[(df['Age'] >= 45) & (df['Age'] <= 60)]

# Calculate median ConvertedComp for each age
median_comp = filtered_df.groupby('Age')['ConvertedComp'].median().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(median_comp['Age'], median_comp['ConvertedComp'], marker='o', linestyle='-', color='green')
plt.title('Median ConvertedComp for Ages 45 to 60')
plt.xlabel('Age')
plt.ylabel('Median ConvertedComp')
plt.xticks(range(45, 61))  # Set x-ticks to show all ages from 45 to 60
plt.grid()

# Display the value at each point
for index, value in enumerate(median_comp['ConvertedComp']):
    plt.text(median_comp['Age'][index], value, round(value, 2), ha='center', va='bottom')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here

import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame
# Count occurrences of each category in MainBranch
main_branch_counts = df['MainBranch'].value_counts()

# Plotting
plt.figure(figsize=(10, 6))
main_branch_counts.plot(kind='barh', color='skyblue')
plt.title('Distribution of Main Branches')
plt.xlabel('Number of Respondents')
plt.ylabel('Main Branch')
plt.grid(axis='x')

# Display the value at the end of each bar
for index, value in enumerate(main_branch_counts):
    plt.text(value, index, value, ha='left', va='center')

plt.show()
```
